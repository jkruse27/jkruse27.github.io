```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

letter = 't'  #Letra analisada
file = "Pronuncia do t - PÃ¡gina1.csv"   #Arquivo contendo as pronuncias
```

```python
a = pd.read_csv(file)
```

```python
sub = a.dropna()

k = lambda x: x[x.find(letter) + 1]
j = lambda x: x[x.find(letter) - 1] if x.find(letter) != 0 else " "

l = sub['Palavra'].map(k).rename("Depois") 
h = sub['Palavra'].map(j).rename("Antes")

sub = pd.concat([l, sub["Pronuncia"]], axis=1, join='inner')
sub = pd.concat([h, sub], axis=1, join='inner')
```

```python
a4_dims = (15, 12)
fig, ax = plt.pyplot.subplots(figsize=a4_dims)

sns.countplot(ax =ax, y='Pronuncia', hue='Depois', data=sub, palette="Spectral").set_title("Pronuncia x Letra Posterior")
```

```python
a4_dims = (15, 12)
fig, ax = plt.pyplot.subplots(figsize=a4_dims)

sns.countplot(ax =ax, y='Pronuncia', hue='Antes', data=sub, palette="Spectral").set_title("Pronuncia x Letra Anterior")
```
